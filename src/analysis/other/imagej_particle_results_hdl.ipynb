```python
from pathlib import Path
from math import pi
import pandas as pd
import plotly.express as px
from scipy.stats import kruskal, mannwhitneyu
```

```python
# Specify input file path
path_input1: Path = Path(
    "../../../data/original/other/imagej_particle_results_hdl1.csv"
).resolve()
path_input2: Path = Path(
    "../../../data/original/other/imagej_particle_results_hdl2.csv"
).resolve()
```

```python
# Define conversion factor: 440 Pixels = 400 nm
cf1: float = 440 / 400
cf2: float = 255 / 50
```

```python
# Read input file
df1: pd.DataFrame = pd.read_csv(path_input1, index_col=0).convert_dtypes()
df2: pd.DataFrame = pd.read_csv(path_input2, index_col=0).convert_dtypes()
```

```python
# Convert pixel values to nm
df1["Area"] = df1["Area"] / (cf1**2)
df1["Perim."] = df1["Perim."] / cf1
df1["Major"] = df1["Major"] / cf1
df1["Minor"] = df1["Minor"] / cf1
```

```python
df2["Area"] = df2["Area"] / (cf2**2)
df2["Perim."] = df2["Perim."] / cf2
df2["Major"] = df2["Major"] / cf2
df2["Minor"] = df2["Minor"] / cf2
```

```python
# Calculate diameter from area
df1["Diameter"] = df1["Area"].apply(lambda x: 2 * (x / pi) ** 0.5)
df2["Diameter"] = df2["Area"].apply(lambda x: 2 * (x / pi) ** 0.5)
```

```python
# Summary statistics
df1[["Area", "Perim.", "Diameter", "Major", "Minor"]].describe()
```

```python
df2[["Area", "Perim.", "Diameter", "Major", "Minor"]].describe()
```

```python
# Plot histograms
fig = px.histogram(df1, x="Area", nbins=100, title="Area (nm)", marginal="box")
fig.show()
```

```python
# Plot histograms
fig2 = px.histogram(df2, x="Area", nbins=100, title="Area (nm)", marginal="box")
fig2.show()
```

### Non-parametric tests

```python
kruskal(df1["Area"], df2["Area"])
```

```python
mannwhitneyu(df1["Area"], df2["Area"], alternative="two-sided")
```
